```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Charger les donn√©es (Les fichiers n'ont pas d'en-t√™te)
# On d√©finit manuellement les noms des 26 colonnes
index_names = ['unit_nr', 'time_cycles']
setting_names = ['setting_1', 'setting_2', 'setting_3']
sensor_names = ['s_{}'.format(i) for i in range(1, 22)]
col_names = index_names + setting_names + sensor_names

# Lecture avec le s√©parateur d'espaces '\s+'
train_df = pd.read_csv('train_FD001.txt', sep='\s+', header=None, names=col_names)

print("--- Phase 1: Donn√©es Charg√©es ---")
display(train_df.head())

# 2. Visualiser l'√©volution des capteurs (Moteur sp√©cifique ex: ID 1)
# On observe comment les capteurs √©voluent jusqu'√† la panne
plt.figure(figsize=(12, 5))
motor_1 = train_df[train_df['unit_nr'] == 1]
plt.plot(motor_1['time_cycles'], motor_1['s_11'], label='Capteur 11', color='red')
plt.plot(motor_1['time_cycles'], motor_1['s_4'], label='Capteur 4', color='blue')
plt.title('√âvolution des capteurs pour le moteur 1 (jusqu\'√† la panne)')
plt.xlabel('Cycles de temps')
plt.ylabel('Valeurs brutes')
plt.legend()
plt.show()

# 3. Question : Tous les capteurs r√©agissent-ils ?
# On identifie les capteurs "plats" (qui n'ont qu'une seule valeur unique)
constant_sensors = [col for col in train_df.columns if train_df[col].nunique() <= 1]
print(f"\nCAPTEURS CONSTANTS IDENTIFI√âS : {constant_sensors}")
print("R√©ponse : Non, ces capteurs ne r√©agissent pas √† la d√©gradation et doivent √™tre supprim√©s.")

# 4. Analyser les corr√©lations (Matrice de corr√©lation)
plt.figure(figsize=(15, 10))
sns.heatmap(train_df.corr(), annot=False, cmap='coolwarm', center=0)
plt.title('Matrice de Corr√©lation des Capteurs')
plt.show()
```

```
from sklearn.preprocessing import MinMaxScaler

# 1. Calcul du RUL (Remaining Useful Life) - Logique du PDF
# Pour chaque moteur (unit_nr), on trouve le cycle maximum atteint
max_cycle = train_df.groupby('unit_nr')['time_cycles'].max().reset_index()
max_cycle.columns = ['unit_nr', 'max_life']

# On fusionne cette info avec le tableau principal
train_df = train_df.merge(max_cycle, on='unit_nr', how='left')

# Formule : RUL = Cycle Max du moteur - Cycle Actuel
train_df['RUL'] = train_df['max_life'] - train_df['time_cycles']
train_df.drop('max_life', axis=1, inplace=True)

# 2. Nettoyage : Supprimer les colonnes inutiles
# On enl√®ve les capteurs constants identifi√©s en Phase 1
train_df.drop(columns=constant_sensors, inplace=True)

# 3. Normalisation (Min-Max Scaling)
# L'IA g√®re mal les grands √©carts (ex: Temp√©rature 600 vs Pression 14)
scaler = MinMaxScaler()

# On s√©lectionne les colonnes √† normaliser (tout sauf ID, cycle et cible RUL)
features_cols = train_df.columns.difference(['unit_nr', 'time_cycles', 'RUL'])
train_df[features_cols] = scaler.fit_transform(train_df[features_cols])

print("\n--- Phase 2: Pr√©traitement Termin√© ---")
print("Nouvelle colonne RUL ajout√©e et capteurs normalis√©s entre [0, 1].")
display(train_df[['unit_nr', 'time_cycles', 'RUL']].head())
```

```
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

# 1. Pr√©paration des Entr√©es (X) et de la Cible (y)
# Directive du PDF : "Ne pas tout donner √† manger √† l'IA"
# On enl√®ve 'unit_nr' et 'time_cycles' car l'IA doit pr√©dire selon l'√âTAT des capteurs, pas selon l'heure
X_train = train_df.drop(columns=['unit_nr', 'time_cycles', 'RUL'])
y_train = train_df['RUL']

print(f"Nombre de caract√©ristiques (features) utilis√©es : {X_train.shape[1]}")

# 2. Approche "Baseline" (R√©f√©rence) : R√©gression Lin√©aire
# C'est un mod√®le simple qui cherche une ligne droite entre les donn√©es
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# 3. Approche Avanc√©e : Random Forest Regressor (For√™ts al√©atoires)
# C'est un mod√®le puissant qui utilise plusieurs "arbres de d√©cision"
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# 4. Comparaison rapide sur les donn√©es d'entra√Ænement
# On regarde si les mod√®les arrivent √† "apprendre" la logique
lr_train_pred = lr_model.predict(X_train)
rf_train_pred = rf_model.predict(X_train)

rmse_lr = np.sqrt(mean_squared_error(y_train, lr_train_pred))
rmse_rf = np.sqrt(mean_squared_error(y_train, rf_train_pred))

print("\n--- Phase 3: Mod√©lisation Termin√©e ---")
print(f"Erreur (RMSE) - R√©gression Lin√©aire : {rmse_lr:.2f}")
print(f"Erreur (RMSE) - Random Forest : {rmse_rf:.2f}")
```
